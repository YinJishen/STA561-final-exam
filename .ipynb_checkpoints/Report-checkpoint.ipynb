## Problem 2

In this problem, I used the whole dataset and created a metric called concentration.

Fisrt of all, I created a matrix $D$ with shape $10\times10$ with $D[i, j]$ as the average distance between all samples with label $i$ and the center (of gravity) of all samples with labels $j$. For a better dimension-deduction method, the diagonal element of $D$ should be smaller than other elements. From this intuition, I define "concentration" as,

$$
\frac{10\sum_{i}D[i,i]}{\sum_{i,j}D[i,j]}
$$

The smaller this value is, the better this method can help distinguish between different digits.

## Problem 3

$$
\begin{aligned}
&\text{E}[(y-\hat{f}(x))^2]\\
=&\text{E}[(f(x)-\hat{f}(x)+\epsilon)^2]\\
=&\text{E}(f(x)-\hat{f}(x))^2+\text{Var}(\epsilon)\\
=&\text{E}(f(x)-\text{E}\hat{f}(x)+\text{E}\hat{f}(x)-\hat{f}(x))^2+\text{Var}(\epsilon)\\
=&(f(x)-\text{E}\hat{f}(x))^2+\text{Var}(\hat{f}(x))+\text{Var}(\epsilon)
\end{aligned}
$$

The first term is the approximation error, the second term is the estimation error, and the third term is the optimization error (irreducilble part).

## Problem 4

Gibbs Sampler for Latent Dirichlet Allocation
    
Step 1: For $i=1$ to $n$: $Z^{(i)}_{(0)}\sim \text{Uni}(1, ..., K)$

Step 2: For $t=1$ to $T$

(a) For each $k,l$, 

sample $P^{(t)}_{(kl)}|X, Z_{(t-1)}\sim \text{Dir}(\lambda_1+n_{kl1},...,\lambda_{J_{l}}+n_{klJ_{l}})$

where $n_{klj}=\sharp\{(i,a):X_{l}^{i,a}=k, Z_{(t-1)}^{(i)}=k\}$

(b) For each i,

$\text{Pr}(Z_{(t)}^{(i)}=k|X, P^{(t)})=\frac{\text{Pr}(X^{(i)}|P^{(t)},z^{(i)}=k)}{\sum_{k'}\text{Pr}(X^{(i,a)}|P^{(t)},z^{(i)}=k')}$

Step 3: (burn in) Remove the first $t_0$ pairs of $(P_{(t)}, Z_{(t)})$

Step 4: (thinning) Keep every $a$-th pair of the remaining $(P_{(t)}, Z_{(t)})$

Step 5: We now have a iid draws from $\text{Pr}(Z, P)$

As we can see, Gibbs Sampler simulates the distribution of target variables by giving pairs of $(P, Z)$, which provides an insight of uncertainty. On the contrary, EM algorithm repeats E-step and M-step until convergence. In each step, we computed an exact result instead of a random number. So we can only get one pair of limitation without the ability of learning the uncertainty of the model.

## Problem 5

1st order optimization is the method that requires at least one first-derivative/gradient, such as BFGS (quasi-Newton Method). 2nd order optimation is any algorithm that uses any second derivative, in the scalar case, such as Newton-Raphson Methods.

Batch optimization is to divide the whole inputs into separate parts randomly, and put all batches into the model one by one. After one epoch, the algorithm iterates over all single inputs. Stochastic optimization is to select inputs randomly in each steps. Such algorithms do not neccesarily iterate over all inputs after one epoch. But the chance of each inputs selected is equal to each other.
